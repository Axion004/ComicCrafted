```python
import os, shutil, pathlib
src = '/content/pytorch-CycleGAN-and-pix2pix/results/facades_label2photo_pretrained/test_latest/images'
dst = '/content/results'
pathlib.Path(dst).mkdir(parents=True, exist_ok=True)
for f in os.listdir(src):
    if f.endswith('_fake_B.png') or f.endswith('_real_B.png') or f.endswith('_real_A.png'):
        shutil.copy(os.path.join(src, f), os.path.join(dst, f))
print('Saved samples to', dst)
```

```python
# Create a clean project folder at /content/task04-pix2pix
import os, shutil, textwrap, pathlib
root = pathlib.Path('/content/task04-pix2pix')
(root / 'results').mkdir(parents=True, exist_ok=True)

# Copy sample results collected earlier
src = pathlib.Path('/content/results')
if src.exists():
    for f in src.glob('*.png'):
        shutil.copy(f, root / 'results' / f.name)

# Write README.md
readme = f"""
# Task-04: Image-to-Image Translation with cGAN (pix2pix)

This project demonstrates pix2pix (conditional GAN) for paired label→photo translation on the Facades dataset using the maintained PyTorch reference codebase.

## Quickstart (Colab)
1) Enable GPU runtime.
2) Install PyTorch, clone the official pix2pix repo, and download the Facades dataset.
3) Option A — Pretrained inference:
   - Download pretrained facades label→photo model.
   - Run test with direction=BtoA and export ~10 samples into results/.
4) Option B — Quick train:
   - Run a short training (e.g., 1 epoch) with model=pix2pix and direction=BtoA.
   - Test to generate samples, then copy images into results/.

## Outputs
- Example inputs/targets/generations are in results/ for report/screenshots.

## Notes
- direction=BtoA is correct for Facades since A→B maps photos→labels.
- For better quality, increase epochs; pretrained weights are provided for deadline-friendly results.

## Credits
- Based on the official PyTorch pix2pix/CycleGAN repository and standard pix2pix practice.
"""
with open(root / 'README.md', 'w', encoding='utf-8') as f:
    f.write(textwrap.dedent(readme))
print('Prepared /content/task04-pix2pix with README and results/')
```
